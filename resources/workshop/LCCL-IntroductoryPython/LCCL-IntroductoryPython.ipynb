## 7. Loops

```
# Looping a set number of times
for i in range(10):
    print('Good evening')

for i in range(0, 10, 2):
    print(i)
```

```
# while-loop
password = input('Enter password: ')

while password != 'ILovePython':
    print('Invalid password.')
    password = input('Enter password: ')

print('Welcome!')
```

```
# iterating through a sequence (more in next workshop)
for char in 'APPLES':
    print(char)
```

```
# break statement
while True:
    password = input('\nEnter password: ')
    if password == 'ILovePython':
        break
    print('Invalid password.')

print('Welcome!\n')
```

```
# continue statement
for i in range(5):
    if i == 2:
        continue 
    print(i)
```

## 8. Functions

```
# defining your own function
def triangle(): 
    for i in range(1, 8):
        print( '*'*i )

# calling your own function
triangle() 
```

```
# function with parameters
def triangle(height): 
    for i in range(1, height+1):
        print( '*'*i )

triangle(3) 
```

```
# return statement
def volume(a, b, c):
  return a*b*c

volume(2, 3, 4)
```

```
# return statement
def triangletype(a, b, c):
  longest = max(a, b, c)
  shortest = min(a, b, c)
  middle = a + b + c - longest - shortest
  
  if longest <= shortest + middle:
    if a == b == c:
      return 'equilateral'
    elif a == b or a == c or b == c:
      return 'isoceles'
    else:
      return 'scalene'
  return 'invalid'

print(triangletype(5, 6, 5))
print(triangletype(5, 6, 7))
print(triangletype(5, 5, 5))
print(triangletype(8, 3, 2))
```
